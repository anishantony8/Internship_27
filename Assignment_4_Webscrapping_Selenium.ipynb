1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#import selenium,webdriver,upload webdriver.exe
from selenium import webdriver as wb
import time
from selenium.webdriver.support.select import Select
#get the url and extract into html
driver = wb.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(10)
from selenium.common.exceptions import NoSuchElementException
S_No,Video_Name,Uploader,Views,Upload_Date = [],[],[],[],[]
for i in driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr'):
    try:
        S_No.append(i.find_elements_by_tag_name('td')[0].text)
        Video_Name.append(i.find_elements_by_tag_name('td')[1].text.split('[')[0])
        Uploader.append(i.find_elements_by_tag_name('td')[2].text)
        Views.append(i.find_elements_by_tag_name('td')[3].text)
        Upload_Date.append(i.find_elements_by_tag_name('td')[4].text)
    except NoSuchElementException as error:
        S_No.append('Not available')
        Video_Name.append('Not available')
        Uploader.append('Not available')
        Views.append('Not available')
        Upload_Date.append('Not available')
import pandas as pd
df_1 = pd.DataFrame(zip(S_No,Video_Name,Uploader,Views,Upload_Date),columns=('S_No','Video_Name','Uploader','Views','Upload_Date'))
df_1
```
